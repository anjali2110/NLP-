## NECESSARY IMPORTS

````python

import spacy
print('spaCy Version: %s' % (spacy.__version__))
spacy_nlp = spacy.load('en_core_web_sm')
````

## STOP WORDS

````python
spacy_stopwords = spacy.lang.en.stop_words.STOP_WORDS
print('Number of stop words: %d' % len(spacy_stopwords))
print('First ten stop words: %s' % list(spacy_stopwords)[:10])
````

````python
with open('dataset') as f:
    text = f.read()
````

## CODE TO REMOVE STOP WORDS FROM THE TEXT

````python
#Remove stop words

article= text

doc = spacy_nlp(article)
tokens = [token.text for token in doc if not token.is_stop]
print('Original Article: %s' % (article))
print()
print(tokens)
````

## CODE TO WRITE CUSTOM STOP WORDS

````python
customize_stop_words = [
    'vit', 'university'
]
for w in customize_stop_words:
    spacy_nlp.vocab[w].is_stop = True
doc = spacy_nlp(article)
tokens = [token.text for token in doc if not token.is_stop]
print('Original Article: %s' % (article))
print()
print(tokens)
````

## NLTK

````python
import nltk 
print('NLTK Version: %s' % (nltk.__version__))
nltk.download('stopwords')
````

## HANDLING STOP WORDS IN NLTK


````python
nltk_stopwords = nltk.corpus.stopwords.words('english')
print('Number of stop words: %d' % len(nltk_stopwords))
print('First ten stop words: %s' % list(nltk_stopwords)[:10])
````

## TOKENIZE OPERATIONS

````python
tokens = nltk.tokenize.word_tokenize(article)
tokens = [token for token in tokens if not token in nltk_stopwords]

print(tokens)
````

## TOKENIZE ALL INCLUDING STOP WORDS

````python
#with everything
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer('\w+|\$[\d\.]+|\S+')
tokenizer.tokenize(article)
````

````python
## TOKENIZE ALL INCLUDING TEXT
````

````python
#only text
tokenizer = RegexpTokenizer('\s+', gaps=True)
tokenizer.tokenize(article)
````

## TOKENIZE UPPER CASE 

````python
#cap_word
capword_tokenizer = RegexpTokenizer('[A-Z]\w+')
tokens=capword_tokenizer.tokenize(article)
````

## PART OF SPEECH TAGGER

````python
nltk.download('averaged_perceptron_tagger')
tagged = nltk.pos_tag(tokens)
tagged[0:6]
````

## CHUNKING


````python
import tkinter
entities = nltk.chunk.ne_chunk(tagged)
entities
````
